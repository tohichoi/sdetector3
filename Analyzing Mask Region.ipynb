```python
data_dir='data/act2/'
mask=cv2.imread('data/' + 'mask1.png')
mask=cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
image=cv2.imread(data_dir + 'frame-000.jpg').astype('uint8')

contours, _ = cv2.findContours(mask, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
s=cv2.contourArea(contours[0])

print(mask.shape)
print(image.shape)
print(s)

mask_region=cv2.bitwise_and(image, image, mask=mask)

fig, ax=plt.subplots(1, 3, figsize=(20, 5))
ax[0].imshow(image)
ax[1].imshow(mask)
ax[2].imshow(mask_region)
fig.show()
```
